This notebook merges the SNPP and NOAA-20 VIIRS detections

```python
import pandas as pd
#pd.set_option("display.max_rows", None, "display.max_columns", None)
from datetime import datetime, timedelta
```

```python
viirs_n20 = pd.read_csv('fire_nrt_J1V-C2_230186.csv')
viirs_snpp = pd.read_csv('fire_archive_SV-C2_230187.csv')
print(viirs_n20.dtypes)
print(viirs_snpp.dtypes)
```

```python
print(viirs_n20.shape, viirs_snpp.shape)
viirs_all = pd.concat([viirs_n20, viirs_snpp], axis = 0)
print(viirs_all.shape)
viirs_all.to_csv('merged_viirs.csv', index=False)
```
